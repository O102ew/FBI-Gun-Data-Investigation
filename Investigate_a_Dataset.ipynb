```python
#to create columns by dividing column over other one the columns must be type of float or int  
```

```python
df_merge_col.dtypes
```

```python
#I convert columns to be float type 
```

```python
df_merge_col.columns
```

```python
df_merge_col.columns=['month_x', 'state', 'totals_x', 'month_y', 'totals_y',
       'Population_2016',
       'Population_2010']
```

```python
df_merge_col.head()
```

```python
df_merge_col.columns.str.strip()
```

```python
#now I remove commas in string in columns "Population_2016" and "Population_2010"
```

```python
df_merge_col["Population_2016"] = df_merge_col["Population_2016"].str.replace(',', '')
df_merge_col.head()
```

```python
df_merge_col["Population_2010"] = df_merge_col["Population_2010"].str.replace(',', '')
df_merge_col.head()
```

```python
df_merge_col["Population_2010"] = df_merge_col["Population_2010"].astype(float)
df_merge_col["Population_2016"] = df_merge_col["Population_2016"].astype(float)
```

```python
df_merge_col["gun_ratio_2010"]=(df_merge_col["totals_x"]/df_merge_col["Population_2010"])*1000
df_merge_col["gun_ratio_2016"]=(df_merge_col["totals_y"]/df_merge_col["Population_2016"])*1000

```

```python
df_merge_col.head(50)
```

```python
#now I draw chart to show the change in gun_ratio_2010 and gun_ratio_2016 over states 
```

```python
df_merge_col.plot(x="state", y=["gun_ratio_2010", "gun_ratio_2016"], kind="barh",figsize=(20,20));
plt.title('gun_ratio in 2010 & 2016');
plt.xlabel('gun_ratio');

```
